```python
import pandas as pd
import numpy as np
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly
from plotly.graph_objs import Scatter, Figure, Layout, Histogram, Box
init_notebook_mode(connected=True)
np.random.seed(68)
%matplotlib inline
```

```python
df = pd.read_csv("rule_ordering.csv")
```

```python
hist_total = Histogram(x=df["time"])
iplot([hist_total])
```

```python
df.head()
```

```python
groups = []
groups.append("Normal")
groups.append("Reverse")
groups.append("Core Number Low to High")
groups.append("Core Number High to Low")
groups.append("Arborescence")
groups.append("Reverse Arborescence")
groups.append("Less Out Edges First")
groups.append("More Out Edges First")
groups.append("Random")
```

```python
df_grp = df[["optim", "time"]].groupby(by="optim")
```

```python
for grp in df_grp.groups:
    print(groups[grp], df_grp.get_group(grp)["time"].mean(), df_grp.get_group(grp)["time"].std())
```

```python
hists = []
for grp in df_grp.groups:
    hists.append(Histogram(x=df_grp.get_group(grp)["time"], name=groups[grp])) #, cumulative=dict(enabled=True)))
iplot(hists)
```

```python
boxs = []
for grp in df_grp.groups:
    boxs.append(Box(y=df_grp.get_group(grp)["time"], name=groups[grp]))
# iplot([boxs[2], boxs[7]])
iplot(boxs)
```
