```python
def show_results(filename):
    df = pd.read_csv(filename)
    grouped = df.groupby(by="n_functions")
    data = []
    means = []
    percentage = []
    for g in grouped.groups:
        means.append((g, grouped.get_group(g).time.get_values().mean()))
        data.append(Histogram(x=grouped.get_group(g).time.get_values(), name=str(g)))
        percentage.append((g, 100.0 * df[df.n_functions == g].size / float(df.size)))
    iplot(data)
    means = sorted(means, key=lambda x: x[0])
    if means[0][0] == 1:
        means[0]=(len(means) + 1, means[0][1])
        means = sorted(means, key=lambda x: x[0])
    percentage = sorted(percentage, key=lambda x: x[0])
    if percentage[0][0] == 1:
        percentage[0]=(len(percentage) + 1, percentage[0][1])
        percentage = sorted(percentage, key=lambda x: x[0])
    x_m = map(lambda x: x[0], means)
    y_m = map(lambda x: x[1], means)
    x_p = map(lambda x: x[0], percentage)
    y_p = map(lambda x: x[1], percentage)
    iplot([Scatter(x=x_m, y=y_m, mode = 'lines')])
    iplot([Scatter(x=x_p, y=y_p, mode = 'lines')])
```

```python
show_results("data/elapsed_time7-5-100-10.log")
```
