```python
df_loop = pd.read_csv("influence_loop2.csv", names=["n_function", "res0", "t0", "rules0",
                                                  "res1", "t1", "rules1",
                                                  "res2", "t2", "rules2",
                                                  "res3", "t3", "rules3",
                                                  "res4", "t4", "rules4",
                                                  "res5", "t5", "rules5"])
```

```python
df_loop.head()
```

```python
df_loop["diff0"] = df_loop["t2"] / df_loop["t0"]
df_loop["diff1"] = df_loop["t3"] / df_loop["t1"]
df_loop["diff2"] = df_loop["t4"] / df_loop["t0"]
df_loop["diff3"] = df_loop["t5"] / df_loop["t1"]
```

```python
df_loop[["diff0", "diff1", "diff2", "diff3"]].head()
```

```python
df_loop.describe()
```

```python
groups = ["No loop, middle rules",
       "No loop, empty rule",
       "One loop, middle rules",
       "One loop, empty rule",
       "Two loop, middle rules",
       "Two loop, empty rule"
      ]
```

```python
layout = Layout(
  #  title='Comparisation Performance With Rule Order Optimization',
    font=dict(family='Latin Modern Math', size=6*3),
    xaxis=dict(
        autorange=True,
        showgrid=False,
        zeroline=False,
        showline=False,
        autotick=True,
        ticks='',
        showticklabels=False
    ),
    yaxis=dict(
        title='Time of Computation (in s)',
        autorange=True,
        showgrid=False,
        zeroline=True,
        showline=True,
        autotick=True,
        ticks='',
        showticklabels=True
    ),
    legend=dict(x=0, y=1)
)
```

```python
boxs = []
for grp in range(6):
    boxs.append(Box(y=df_loop["t" + str(grp)], name=groups[grp]))
# iplot([boxs[2], boxs[7]])
fig = Figure(data=boxs, layout=layout)
iplot(fig)
```
